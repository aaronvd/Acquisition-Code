# **Measure SOL Calibration - One Port**

```python
import skrf as rf
from skrf.vi.vna import PNA
from skrf import Network
import matplotlib.pyplot as plt
from pathlib import Path
from datetime import date
import time

rf.stylely()
```

### Build folder for cal files

```python
today = date.today()
date_str = today.strftime("%Y-%m-%d")
cal_folder_name = date_str + '_SOL'

root_folder = Path('C:/Users/lab/Documents/Aaron/DARPA/Data/Aaron_Quick_Turn_Antenna/Test_Board_21-10-07/')
meas_folder = root_folder / cal_folder_name

meas_folder.mkdir(exist_ok = True)
```

### Initialize VNA

```python
f_start = 5E9
f_stop = 15E9
f_npoints = int(801)

vna = PNA(address ='TCPIP0::10.236.73.132::inst0::INSTR')
time.sleep(3)
vna.reset()
vna.scpi.set_trigger_manual

meas_channel = 2

meas_S11 = 'S11'
vna.create_meas(meas_S11,'S11', channel = meas_channel) 

vna.set_frequency_sweep(f_start, 
                        f_stop, 
                        f_npoints, 
                        f_unit = 'Hz',
                        channel = meas_channel, 
                        )
```

### Take measurements

#### *Short S11*

```python
vna.sweep
measurement = vna.get_measurement(mname = meas_S11)
print(f'{meas_S11} measured')

filename = meas_folder / ('SHORT_' + meas_S11 + '.ntwk')
rf.write(str(filename), measurement)
measurement.frequency.units = 'ghz'
measurement.plot_s_db(label ='Short S11')
```

#### *Open S11*

```python
vna.sweep
measurement = vna.get_measurement(mname = meas_S11)
print(f'{meas_S11} measured')

filename = meas_folder / ('OPEN_' + meas_S11 + '.ntwk')
rf.write(str(filename), measurement)

measurement.frequency.units = 'ghz'
measurement.plot_s_db(label ='Open S11')
```

#### *Load S11*

```python
vna.sweep
measurement = vna.get_measurement(mname = meas_S11)
print(f'{meas_S11} measured')

filename = meas_folder / ('LOAD_' + meas_S11 + '.ntwk')
rf.write(str(filename), measurement)

measurement.frequency.units = 'ghz'
measurement.plot_s_db(label ='Load S11')
```
